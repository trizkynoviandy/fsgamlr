```python
import fsgamlr
import pandas as pd
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
X = dataset.data
y = dataset.target
```

```python
fsgamlr = fsgamlr.GeneticAlgorithm(X, y, 
                                   max_features=5, 
                                   population_size=100,
                                   n_generation=50)
```

```python
result = fsgamlr.optimize(verbose=0)
```

```python
fsgamlr.plot_result()
```
