```python
import random
import pandas as pd

from fsgamlr import GeneticAlgorithm
from sklearn.datasets import load_diabetes

random.seed(1)
```

```python
dataset = load_diabetes()
```

```python
X = dataset.data
y = dataset.target
```

```python
fsgamlr = GeneticAlgorithm(X, y, 
                                   max_features=4, 
                                   population_size=200,
                                   n_generation=50)
```

```python
result = fsgamlr.optimize(verbose=0)
```

```python
fsgamlr.plot_result()
```
